```python
#create a DataFrame from the tweet_json.txt file
tweet_data = []
with open('tweet_json.txt', 'r') as tweet_df:
    for line in tweet_df:
        twitter_data = json.loads(line)
        tweet_id = twitter_data['id']
        favorite_count = twitter_data['favorite_count']
        retweet_count = twitter_data['retweet_count']
        tweet_data.append({'tweet_id' : tweet_id, 'favorite_count' : favorite_count, 'retweet_count' : retweet_count})
twitter_data = pd.DataFrame(tweet_data, columns = ['tweet_id', 'favorite_count', 'retweet_count'])
twitter_data
```

### Issue 3:

### Define: Create a column for dogs breeds

### Code

```python
twitter_archive_clean['dog_breed'] = ''

for i, row in twitter_archive_clean.iterrows():

    if row.p1_dog:
        twitter_archive_clean.set_value(i, 'dog_breed', row.p1)
    elif row.p2_dog and row.rating_numerator >= 10:
        twitter_archive_clean.set_value(i, 'dog_breed', row.p2)
    elif row.p3_dog and row.rating_numerator >= 10:
        twitter_archive_clean.set_value(i, 'dog_breed', row.p3)
    else:
        twitter_archive_clean.set_value(i, 'dog_breed', 'None')
```

```python
#Plot a Histogram to visualize dog breeeds
dog_breed = twitter_archive_clean_master.groupby('dog_breed').filter(lambda x: len(x) >= 4)

dog_breed['dog_breed'].value_counts().plot(kind = 'barh')
plt.title('Most popular dog breed')
plt.xlabel('Count')
plt.ylabel('Dog breeds');
```

```python
twitter_archive_clean_master.timestamp = pd.to_datetime(twitter_archive_clean_master['timestamp'], format='%Y-%m-%d %H:%M:%S.%f')

monthly_tweets = twitter_archive_clean_master.groupby(pd.Grouper(key = 'timestamp', freq = "M")).count().reset_index()
monthly_tweets = monthly_tweets[['timestamp', 'tweet_id']]
monthly_tweets.head()
monthly_tweets.sum()
```

```python
# Make a line plot of time vs. tweets
import datetime
plt.figure(figsize=(10, 10));
plt.xlim([datetime.date(2015, 11, 30), datetime.date(2017, 7, 30)]);

plt.xlabel('Year and Month')
plt.ylabel('Tweets Count')

plt.plot(monthly_tweets.timestamp, monthly_tweets.tweet_id);
plt.title('Tweets over Time');
```

```python
# Scatterplot of retweets vs favorite count

sns.lmplot(x="retweet_count", 
           y="favorite_count", 
           data=twitter_archive_clean_master,
           size = 5,
           aspect=1.3,
           scatter_kws={'alpha':1/5});

plt.title('Favorite Count vs. Retweet Count');
plt.xlabel('Retweet Count');
plt.ylabel('Favorite Count');
```
