### WeRateDog Report

### 1: Most popular dog breed
Plotting a bar chart of the dog breeds shows that the most popular breed of dog is the Golden_Retriever followed by the Labrador_Retriever.

### 2. Tweets over Time
This analysis shows the relationship between tweets collected over time. The chart below shows a decline in the number of tweets, from 2016-01 there was a sharp decline. The least tweet collected over the period under review were recorded in 2016-09 and 2017-05 respectively.

### 3. Favorite Count vs. Retweet Count
Analysis shows that there is a positive correlation between the number of retweets and the favourite counts(number of likes).This means that as the number of retweets increases, the number of likes also increases.
